# Plotting with Pandas
* [Source](https://realpython.com/pandas-plot-python/)

```python
import pandas as pd

# to display plot inline 
%matplotlib inline

# to display plots in separate window
# %matplotlib qt
import matplotlib.pyplot as plt
```

## Load Data

```python
download_url = ("https://raw.githubusercontent.com/fivethirtyeight/"
                "data/master/college-majors/recent-grads.csv")

df = pd.read_csv(download_url)
df.head()
```

```python
# make pandas diaplay all columns
pd.set_option("display.max.columns", None)
df.head()
```

## Line plot

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
```

```python
df.plot(x="Rank", y="P75th")
```

## Distributions and Histograms

```python
# histogram
df["Median"].plot(kind="hist")
```

The histogram shows the data grouped into ten bins ranging from \\$20,000 to \\$120,000, and each bin has a width of $10,000. The histogram has a different shape than the normal distribution, which has a symmetric bell shape with a peak in the middle.

```python
top_5 = df.sort_values(by="Median", ascending=False).head()
top_5.plot(x="Major", y="Median", kind="bar", rot=5, fontsize=4)
```

```python
top_medians = df[df["Median"] > 60000].sort_values("Median")
top_medians.plot(x="Major", y=["P25th", "Median", "P75th"], kind="bar")
```

## Scatter Plot

```python
df.plot(x="Median", y="Unemployment_rate", kind="scatter")
```

## Grouping Categorical Data

```python
cat_totals = df.groupby("Major_category")["Total"].sum().sort_values()
cat_totals
```

### Barh

```python
cat_totals.plot(kind="barh", fontsize=4)
```

### Pie Chart

```python
cat_totals.plot(kind="pie", fontsize=4)
```

## Zooming in on Categories

```python
df[df["Major_category"] == "Engineering"]["Median"].plot(kind="hist")
```
